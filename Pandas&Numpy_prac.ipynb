# **Covariance and correlation**

```
import numpy as np
import seaborn as sns
sns.set()
def de_mean(x):
  xmean =  np.mean(x)
  return [xi - xmean for xi in x]

def covariance(x,y):
  n = len(x)
  return np.dot(de_mean(x) , de_mean(y)) / (n-1)

pagespeeds = np.random.normal(3.0 ,1.0 ,1000)
purchaseAmount = np.random.normal(50 , 10.0 ,1000)
sns.scatterplot(pagespeeds , purchaseAmount)
covariance(pagespeeds, purchaseAmount)
```

```
purchaseAmount = np.random.normal(50 , 10 , 1000) / pagespeeds
sns.scatterplot(pagespeeds , purchaseAmount)

covariance (pagespeeds , purchaseAmount)
```

# **Conditional Probability**

```
from numpy import  random
random.seed(0)
totals = {20:0 , 30:0 , 40:0 , 50:0 , 60:0 , 70:0}
purchases = {20:0 , 30:0 , 40:0 , 50:0 , 60:0 , 70:0}
totalPurchases = 0 
for i in range(10000):
  ageDecade = random.choice([20,30,40,50,60,70])
  purchaseProbability = float(ageDecade)/100.0
  totals[ageDecade] += 1
  if (random.random() < purchaseProbability):
    totalPurchases += 1
    purchases[ageDecade] +=1
```

```
purchases
```

```
totalPurchases
```

```
PEF = float(purchases[30] / float(totals[30]))
PEF
```

# **Linear Regression**

```
import numpy as np
import seaborn as sns
sns.set()
pagespeeds = np.random.normal(3.0 , 1.0 ,1000)
purchaseAmount = 100 - (pagespeeds + np.random.normal(0 , 0.1 , 1000)) *3 
sns.scatterplot(pagespeeds , purchaseAmount)
```

```
from scipy import stats
slope , intercept , r_value , p_value , std_err = stats.linregress(pagespeeds , purchaseAmount)

r_value ** 2
```

```
def predict(x):
  return slope * x + intercept
fmri = sns.load_dataset("fmri")
fitline = predict(pagespeeds)
sns.scatterplot(pagespeeds , purchaseAmount)
sns.lineplot(pagespeeds , fitline , color = 'red')
```

# **Polynomial Regression**

```
np.random.seed(2)
pagespeeds = np.random.normal(3.0 ,1.0 , 1000)
purchaseAmount = np.random.normal(50.0, 10.0 , 1000) / pagespeeds
sns.scatterplot(pagespeeds , purchaseAmount , color='purple')
```

```
x = np.array(pagespeeds)
y = np.array(purchaseAmount)
p4 = np.poly1d(np.polyfit(x,y,4))
```

```
xp = np.linspace(0 , 7 ,100)
sns.scatterplot(x , y )
sns.lineplot(xp , p4(xp) , color='red')
```

```
from sklearn.metrics import r2_score

r = r2_score(y , p4(x))
r
```
